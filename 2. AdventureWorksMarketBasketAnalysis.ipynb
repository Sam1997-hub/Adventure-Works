```python
import pandas as pd
import numpy as np
```

```python
sales_2020=pd.read_csv('AdventureWorks Sales Data 2020.csv')
```

```python
sales_2021=pd.read_csv('AdventureWorks Sales Data 2021.csv')
```

```python
sales_2022=pd.read_csv('AdventureWorks Sales Data 2022.csv')
```

```python
product=pd.read_csv('AdventureWorks Product Lookup.csv')
```

```python
sales=pd.concat([sales_2020,sales_2021,sales_2022],ignore_index=True,axis=0)
```

```python
sales.isnull().sum()
```

```python
product.isnull().sum()
```

```python
sales=sales.merge(product,how='inner',on='ProductKey')
```

```python
sales=sales[sales['OrderQuantity']>0]
```

```python
analyse=sales.groupby(['OrderNumber','ProductDescription'])['OrderQuantity'].sum().unstack().fillna(0).reset_index().set_index('OrderNumber')
```

```python
def encode_units(x):
    if (x<=0):
        return 0
    else:
        return 1


basket_plus= analyse.applymap(encode_units)
basket_plus
        
```

```python
basket_encode=basket_plus[(basket_plus>0).sum(axis=1)>=2]
```

```python
basket_encode
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_item_plus=apriori(basket_encode,min_support=0.03,use_colnames=True).sort_values('support',ascending=False).reset_index(drop=True)
```

```python
freq_item_plus['Number of Items']=freq_item_plus['itemsets'].apply(lambda x:len(x))
```

```python
freq_item_plus
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(freq_item_plus,metric='lift',min_threshold=1).sort_values('lift',ascending=False).reset_index()
```
